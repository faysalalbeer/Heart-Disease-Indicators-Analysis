```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/heart_2022_with_nans.csv')
df
```

```python
missing_cols = df.columns[df.isnull().any()].tolist()
col_types = []
examples = []
missing_counts = []

for col in missing_cols:
    sample_vals = df[col].dropna().unique()[:10]
    examples.append(sample_vals)
    missing_counts.append(df[col].isnull().sum())
    if pd.api.types.is_numeric_dtype(df[col]):
        col_types.append("numeric")
    else:
        col_types.append("categorical")


missing_col_summary = pd.DataFrame({
    'Column': missing_cols,
    'Type': col_types,
    'Sample_Values': examples,
    'Missing_Count': missing_counts
})
missing_col_summary = missing_col_summary.sort_values("Missing_Count", ascending=False).reset_index(drop=True)
missing_col_summary_before = missing_col_summary.copy()
missing_col_summary_before

```

```python
df['TetanusLast10Tdap'] =  df['TetanusLast10Tdap'].fillna('No')
df['PneumoVaxEver'] = df['PneumoVaxEver'].fillna('No')
df['HIVTesting'] = df['HIVTesting'].fillna('No')
df['ChestScan'] = df['ChestScan'].fillna('No')
df['CovidPos'] = df['CovidPos'].fillna('No')
df['HighRiskLastYear'] = df['HighRiskLastYear'].fillna('No')
df['FluVaxLast12'] = df['FluVaxLast12'].fillna('No')
df['AlcoholDrinkers'] = df['AlcoholDrinkers'].fillna('No')
df['ECigaretteUsage'] = df['ECigaretteUsage'].fillna('No')
df['SmokerStatus'] = df['SmokerStatus'].fillna('No')
df['GeneralHealth'] = df['GeneralHealth'].fillna(df['GeneralHealth'].mode()[0])
df['LastCheckupTime'] = df['LastCheckupTime'].fillna(df['LastCheckupTime'].mode()[0])
df['RemovedTeeth'] = df['RemovedTeeth'].fillna(df['RemovedTeeth'].mode()[0])
df['RaceEthnicityCategory'] = df['RaceEthnicityCategory'].fillna(df['RaceEthnicityCategory'].mode()[0])
df['AgeCategory'] = df['AgeCategory'].fillna(df['AgeCategory'].mode()[0])
col_to_fill = ['DifficultyErrands', 'DifficultyConcentrating', 'DifficultyWalking',	'DifficultyDressingBathing', 'BlindOrVisionDifficulty',	'DeafOrHardOfHearing',
               'HadAngina',	'HadSkinCancer',	'HadHeartAttack',	'HadDepressiveDisorder', 'HadArthritis', 'HadCOPD', 'HadKidneyDisease',
               'HadAsthma', 'HadStroke',	'PhysicalActivities', 'HadDiabetes']
for col in col_to_fill:
    df[col] = df[col].fillna('No')

```

**For Numeric Data**

```python
mask = df['BMI'].isna() & (df['HeightInMeters'].notna() & df['WeightInKilograms'].notna())
calc_bmi = df.loc[mask, 'WeightInKilograms'] / (df.loc[mask, 'HeightInMeters'] **2)
df.loc[mask, 'BMI'] = calc_bmi
df['BMI'] = df['BMI'].round(2)
x= ["BMI", "WeightInKilograms", "HeightInMeters", "SleepHours"]
for col in x:
    df[col] = df[col].fillna(df[col].median())

df['PhysicalHealthDays'] = df['PhysicalHealthDays'].fillna(0)
df['MentalHealthDays'] = df['MentalHealthDays'].fillna(0)
```
